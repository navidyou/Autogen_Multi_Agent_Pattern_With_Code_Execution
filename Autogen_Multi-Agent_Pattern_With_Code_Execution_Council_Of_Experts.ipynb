```python
import autogen
from autogen import ConversableAgent
from autogen import GroupChat
from autogen import GroupChatManager


llm_config = {"model": "gpt-4o",
              "api_key": "YOUR_API_KEY"}
```

```python
Agent_function = "An idea is given to you. Your role is to advocate for the feasibility of the idea and provide a well-structured, realistic action plan. Use all resources available to you to support your arguments. There is a user proxy that can execute the codes that you werite.You could write code to scrape the web to extract information and content of articles. You could write code to perform simulations,to solving mathematical problems, make plots to support your arguments. Present a comprehensive and practical action plan, addressing potential challenges and demonstrating how they can be overcome. Your goal is to ensure the final outcome is a robust, evidence-based action plan that can be presented to regulators."
```

```python
political_advisor = ConversableAgent(
    name="political_advisor",
    system_message=(
        "You are a political advisor. " + Agent_function
    ),
    llm_config=llm_config,
)
```

```python
political_advisor_user_proxy = autogen.UserProxyAgent(
    name="political_advisor_user_proxy",
    human_input_mode="NEVER",
    max_consecutive_auto_reply=4,
    is_termination_msg=lambda x: x.get("content", "").rstrip().endswith("POLITICAL_ADVISOR_USER_PROXY_TERMINATE"),
    code_execution_config={
        "work_dir": "web",
        "use_docker": False,
    },  # Please set use_docker=True if docker is available to run the generated code. Using docker is safer than running the generated code directly.
    llm_config=llm_config,
    system_message="""Reply POLITICAL_ADVISOR_USER_PROXY_TERMINATE if the task has been solved at full satisfaction.
Otherwise, reply CONTINUE, or the reason why the task is not solved yet.""",
)
```

```python
political_advisor_group_chat = GroupChat(
    agents=[political_advisor, political_advisor_user_proxy],
    messages=[],
    max_round=4,
    speaker_selection_method ="round_robin",
)
political_advisor_group_chat_manager = GroupChatManager(
    groupchat=political_advisor_group_chat,
    llm_config={"config_list": [llm_config]},
)
```
